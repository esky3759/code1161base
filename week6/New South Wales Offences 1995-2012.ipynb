# Investigating NSW Criminal Offences between 1995-2012

# Imports

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
```

```python
%matplotlib inline
plt.rcParams['figure.figsize'] = (20, 10)
```

```python
saved_style_state = matplotlib.rcParams.copy() #give us a style state to go back to
```

```python
print os.path
if os.path.isfile("RCI_OffenceByMonth_1995-2012.csv"):
    filepath = "downloads/RCI_OffenceByMonth_1995-2012.csv"
    print "loading from file"
else:
    filepath = "http://data.gov.au/storage/f/2013-09-12T23%3A32%3A36.918Z/rci-offencebymonth.csv"
    print "loading from the internet"
RCI_OffenceByMonth_9512 = pd.read_csv(filepath)
print "done"
```

```python
# Getting rid of all whitespace in months-years
RCI_OffenceByMonth_9512.columns = [x.strip().replace(" ", "_") for x in RCI_OffenceByMonth_9512.columns]
test_df = RCI_OffenceByMonth_9512
```

# Making each column a heading

```python
test_df.columns
```

# Investigating Offences in 1995

```python
#Taking columns Jan_1995 to Dec_1995 as they make up the year of 1995
years95 = test_df.iloc[:, 4:16]
years95['LGA'] = test_df.iloc[:, 1]
years95
```

```python
#Summing up all crimes in each area of each month in 1995
muppet95 = years95.groupby('LGA').sum()
muppet95
```

```python
#Summing up the total of all months in each area
clown95 = muppet95.sum(axis=1)
clown95
```

```python
#Sorting each area in values from highest to lowest
newcolumnssorted95 = clown95.sort_values(ascending=False)
newcolumnssorted95
```

```python
#Collecting the Top Ten criminal areas in 1995
topten95 = newcolumnssorted95.iloc[:10]
topten95
```

```python
#Bar chart of the Top Ten Criminal Areas in 1995
topten95graph = topten95.plot.bar(x='LGA', y='Jan_1995', facecolor = 'cyan')
plt.title('Top 10 Criminal Areas 1995', fontsize=20)
plt.xlabel('Cities', fontsize=16)
plt.ylabel('Number of Crimes', fontsize=16)
topten95graph
```
